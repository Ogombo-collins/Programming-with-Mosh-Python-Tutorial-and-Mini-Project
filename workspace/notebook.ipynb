## Constructor

A constructor is a function that gets called at the time of creating an object.

```python
#create Pointer class using a constructor

class Pointer:
    def __init__(self, x, y): #x and y are attributes of objects created from Pointer class
        self.x = x
        self.y = y
    def move(self):  #the first method of the Pointer class
        print("move")
    
    def draw(self):  #second method of Pointer class
        print("draw")
```

```python
#instatiate object for Pointer class

pointer1 = Pointer(10, 20)

print(pointer1.x, pointer1.y)
```

```python
#create Person class with name attribute and talk method

class Person:
    """ class Person that has one atttribute; name attribute and one method;talk
    params:
        name(str): name of object created from class
        talk(): function describing action of talking
    """
    def __init__(self, name):
        self.name = name
        
    def talk(self):
        print(f"Hi, i am {self.name}")
```

```python
#instiate Person object

collins = Person("Ogombo")
collins.name
```

```python
#Apply talk method on collins object
collins.talk()
```

```python
#add another object of Person class
mercy = Person("Ethan Hunt")
print(mercy.name)
mercy.talk()
```

## Inheritance

Inheritance is a mechanism for reusing code.

```python
#create mammal class

class Mammal:
    def __init__(self, name):
        self.name = name
    
    def walk(self):
        print("It walks")
        
#instatiate two classes; cat and dog. They will inherit the attribute and method of Mammal class

class Dog(Mammal):
    pass

class Cat(Mammal):
    pass
```

```python
#instatiate  objects of Dog and Cat class
Dog1 = Dog("kate")
print(Dog1.name)
Dog1.walk()


cat1 = Cat("dazzly")
print(cat1.name)
```

## Project 1: Automation with Python

**Goal of the project**:
- Write a Python program that can process thousands of spreadsheets in 4 seconds.


**Value of Project**:
- Saves time in updating thoudands of spreadsheets. Efficient time usage affects persoel prodductivity, thus impacting output that is directly or indirectly linked to company bottomline. 

**Project Methodology**:
1. import openpyxl package: cotains all functions required to work with xlsx format files
2. load dataset into working environment: dataset is xlsl format; transactions.xlsx
3. Load and access workbook of the xlsx file
4. Access contents of worksheet
5. Update worksheet by adding a new column
6. Add charts detailing infomation about the new column
7. Save workbook
8.  Wrap step3-step 7 into a function

```python
import openpyxl as xl
from openpyxl.chart import BarChart, Reference
```

```python
#load dataset
workbook = xl.load_workbook('transactions.xlsx')

#access contents of the worksheet in the workbook loaded into environment
worksheet = workbook['Sheet1']
worksheet
```

```python
#access contents of cell 1 in row 1
cell = worksheet.cell(1,1)

cell_value = cell.value #value of the cell

#print value of the cell
print(cell_value)
```

```python
#number of rows in worksheet
no_rows = worksheet.max_row

print(f"number of rows in worksheet: {no_rows}")
```

```python
#iterate over all rows and determine values in third column, ignoring the header row of the worksheet

for row in range(2, no_rows + 1): #up to fourth row, ignore header row
    cell_values = worksheet.cell(row, 3)
    print(cell_values.value)
```

```python
# Convert worksheet to ascertain if values iterated are similar to the ones in the actual worksheet

import pandas as pd

df = pd.DataFrame(worksheet.values)

df.head()
```

```python
# Add a new column in worksheet

for row in range(2, no_rows + 1): #up to fourth row, ignore header row
    cell_value = worksheet.cell(row, 3).value # Retrieve the value of the cell
    updated_price = cell_value * 0.9
    updated_price_cell = worksheet.cell(row, 4) #adding a new column
    updated_price_cell.value = updated_price
   
#add charts/plotsabs
values = Reference(worksheet, 
          min_row=2, max_row=worksheet.max_row,
         min_col=4,
         max_col=4)

#instance of BarChart class
chart = BarChart()
chart.add_data(values) #adding data to chart
worksheet.add_chart(chart, "e2") #adding chart to the worksheet
    
#save workbook to new file with updated changes
workbook.save("transactions_updated.xlsx")
```

## Automating Function

Based on the output above, it's clear the program written to automate updating of thousands of xlsx files works.

To make sure, the program can be used on multiple xlsx files, we wrap the script into a function as shown below.

```python
#define function that automates the process

def process_workbook(filename):
    wb = xl.load_workbook(filename)
    sheet = wb["sheet1"]
    
    # Add a new column in worksheet

    for row in range(2, no_rows + 1): #up to fourth row, ignore header row
        cell_value = sheet.cell(row, 3).value # Retrieve the value of the cell
        updated_price = cell_value * 0.9
        updated_price_cell = sheet.cell(row, 4) #adding a new column
        updated_price_cell.value = updated_price

    #add charts/plotsabs
    values = Reference(sheet, 
              min_row=2, max_row=sheet.max_row,
             min_col=4,
             max_col=4)

    #instance of BarChart class
    chart = BarChart()
    chart.add_data(values) #adding data to chart
    sheet.add_chart(chart, "e2") #adding chart to the worksheet

    #save workbook to new file with updated changes
    wb.save(filename)
```
